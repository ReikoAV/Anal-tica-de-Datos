```python
#Este es un comentario de prueba 
import numpy as np
from matplotlib import pyplot as plt
ys = 200 + np.random.randn(100)
x = np.arange (0, 100, 1)
```

```python
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, facecolor='paleturquoise',alpha=0.5)
plt.title('Sample Visualization',size=16)
plt.xlabel('Eje de las X',size=14)
plt.ylabel('Eje de las Y',size=14)
plt.show
```
